```python
# if/elif/else — same logic as SQL CASE WHEN
score = 78

if score >= 90:
    grade = "A"
elif score >= 75:
    grade = "B"
elif score >= 60:
    grade = "C"
else:
    grade = "Fail"

print(f"Score {score} → Grade: {grade}")
```

```python
sales = [12000, 8500, 22000, 6000, 18500]

total = 0
for amount in sales:
    total = total + amount
    print(f"  Added {amount:,}  →  Running total: {total:,}")

print(f"\nFinal total: {total:,}")
print(f"Average:     {total/len(sales):,.0f}")
```

```python
# List comprehension: shorter version of a for loop that builds a list
high_sales = [x for x in sales if x > 10000]
print("Sales above 10k:", high_sales)
```

```python
monthly_sales = [45000, 12000, 88000, 6500, 34000, 71000, 9200]

for month_num, amount in enumerate(monthly_sales, start=1):
    if amount >= 70000:
        tier = "🔥 High"
    elif amount >= 30000:
        tier = "✅ Medium"
    else:
        tier = "⚠️  Low"
    print(f"Month {month_num:2d}: ${amount:6,}  → {tier}")
```

```python
def calculate_margin(revenue, cost):
    """
    Returns gross margin as a percentage.
    Example: calculate_margin(100, 60) → 40.0
    """
    if revenue == 0:
        return 0
    return (revenue - cost) / revenue * 100

# Call the function with different values
print(calculate_margin(250000, 65000))  # → 35.0
print(calculate_margin(50000, 50000))    # → 0.0 (no profit)
print(calculate_margin(0, 0))            # → 0 (avoids division by zero)
```

```python
def calculate_margin(revenue, cost):
    """
    Returns gross margin as a percentage (1 decimal place).
    """
    if revenue == 0:
        return 0
    return round((revenue - cost) / revenue * 100, 1)

# Call the function with different values
print(calculate_margin(150000, 65000))  # → 74.0
print(calculate_margin(50000, 50000))   # → 0.0
print(calculate_margin(0, 0))           # → 0
```

```python
def calculate_margin(revenue, cost):
    """
    Returns gross margin as a percentage (1 decimal place).
    """
    if revenue == 0:
        return 0
    return f"{(revenue - cost) / revenue * 100:.0f}"

# Call the function with different values
print(calculate_margin(250000, 65000))  # → 74.0
print(calculate_margin(50000, 50000))   # → 0.0
print(calculate_margin(0, 0))           # → 0
```

```python
def classify_revenue(amount, low_threshold=20000, high_threshold=60000):
    """
    Classifies a revenue amount as Low, Medium or High.
    Default thresholds can be overridden when calling the function.
    """
    if amount >= high_threshold:
        return "High"
    elif amount >= low_threshold:
        return "Medium"
    else:
        return "Low"

# Using defaults
print(classify_revenue(75000))   # → High
print(classify_revenue(35000))   # → Medium
print(classify_revenue(5000))    # → Low

# Overriding defaults for a different business context
print(classify_revenue(35000, low_threshold=10000, high_threshold=50000))
```

```python
def calculate_margin(revenue, cost):
    return round(((revenue - cost) / revenue) * 100,1)

def classify_revenue(revenue):
    if revenue >= 70000:
        return "High"
    elif revenue >= 30000:
        return "Medium"
    else:
        return "Low"

transactions = [
    {"product": "Laptop",  "revenue": 85000, "cost": 55000},
    {"product": "Mouse",   "revenue": 12000, "cost": 6000},
    {"product": "Monitor", "revenue": 42000, "cost": 28000},
]

for t in transactions:
    margin = calculate_margin(t["revenue"], t["cost"])
    tier   = classify_revenue(t["revenue"])
    print(f"{t['product']} | Revenue Tier: {tier} | Margin: {margin:.1f}%")
```

```python
def top_n(items, n=3):
    """Returns the top N items from a list, sorted descending."""
    return sorted(items, reverse=True)[:n]

sales_amounts = [5000, 82000, 14000, 63000, 9000, 47000, 31000]
print("Top 3 sales:", top_n(sales_amounts))
print("Top 5 sales:", top_n(sales_amounts, n=5))
```
